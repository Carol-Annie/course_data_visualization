```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
categorias = ['A', 'B', 'C', 'D']
valores = [10,20,15,5]
#tupla de largura e altura
plt.figure(figsize=(9,3))
#coloração
plt.bar(categorias,valores,color='lightseagreen')
plt.xlabel('Grupos')
plt.ylabel('Tamanho')
plt.title('Gráfico em barra de categorias')
plt.show()
```

```python
url_data = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(url_data)
filmes.head(15)
```

```python
#antos filmes existem de cada gênero?
contagens = filmes.groupby(by='Genre').count()
contagens
```

```python
from matplotlib import colorbar
plt.bar(contagens.index, contagens['Film'],color='darkseagreen')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,5))
plt.title('Quantidade de filmes por gênero')
sns.countplot(y="Genre", data=filmes)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Bilheteria média por gênero')
sns.barplot(x="Genre", y="Worldwide_Gross_M", color = 'navajowhite', ci = 'sd', data=filmes)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Bilheteria média por gênero')
sns.barplot(x="Genre", y="Worldwide_Gross_M", color = 'darksalmon', estimator= sum ,ci = None, data=filmes)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('Filmes por gênero em cada ano')
sns.countplot(x="Year", hue= "Genre", data=filmes)
plt.show()


plt.figure(figsize=(15,8))
plt.title('Média de bilheteria por gênero em cada ano')
sns.countplot(x="Year", hue= 'Genre', palette = "pastel", edgecolor="0.4", data=filmes)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
movies_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt'
movies = pd.read_json(movies_uri)

plt.figure(figsize=(15,5))
plt.title('Orçamento Filmes')
sns.barplot(x="Film", y="Budget_M", color = 'mediumaquamarine' ,ci = None, data=movies)
plt.xticks(fontsize=15,rotation=90)
plt.show()

'''
Maior orçamento: Pirates of the Caribbean: At World's End
Menor orçamento: The Twilight Saga: New Moon
'''
```

```python
plt.figure(figsize=(15,5))
plt.title('Bilheteria Filmes')
sns.barplot(x="Film", y="Worldwide_Gross_M", color = 'mediumseagreen' ,ci = None, data=movies)
plt.xticks(fontsize=15,rotation=90)
plt.show()

'''
Maior bilheteria: Avatar
Menor bilheteria: Iron Man 2, Hancok, Fast Five, Kung Fu Panda e Ratatouille
'''
```

```python
plt.figure(figsize=(15,8))
plt.title('Filmes por gênero em cada ano')
sns.countplot(x="Year", hue= "Genre", palette = "muted", data=movies)
plt.show()

'''
Menos filmes de ação: 2008
Menos filmes de ação animação: 2008 e 2011
'''
```

```python
plt.figure(figsize=(10,5))
plt.title('Orçamento médio por gênero')
#sns.countplot(x="Year", hue= "Budget_M", palette = "muted", data=movies)
sns.barplot(x="Genre", y="Budget_M", color= 'plum', ci = 'sd', data=movies)
plt.show()
```
